```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df = pd.read_csv('Feo_analysis_data.csv')
df.set_index('Date',drop=True, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(11,7),dpi=80)
plt.title('Grate Speed (rpm)', fontsize=20,fontweight='bold')
f=sns.lineplot(x='Date', y=df['Grate_Speed(rpm)'], data=df,)
f.set_xlabel("Date", fontsize = 18)
f.set_ylabel('Grate Speed', fontsize = 18)
f.set_xticks([1,200,400,550,645])
plt.show()
```

```python
data = df['Pellet_Feo'].value_counts()
data
```

```python
plt.figure(figsize=(11,7))
plt.text(1,1,'Feo < 0.5%                ==> Very Good')
plt.text(1,1.1,'0.5% < Feo < 0.8%   ==> Good')
plt.text(1,1.2,'0.8% < Feo < 1.2%   ==> Not Bad')
plt.text(1,1.3,'Feo > 1.2%                ==> Bad')
data.plot.pie(autopct = '%0.0f%%', textprops={'fontsize': 18})
```

```python
sns.displot(
    data=df,
    x="TG_Zone1_Temp(deg)",
    hue="Pellet_Feo",
    hue_order=['Very Good','Good','Not Bad','Bad'],
    kind="kde", height=7,
    fill=True,
    alpha = 0.05,
)
plt.xlim(190,260)
plt.xticks([210,236,224])
plt.axvline(236, c='green')
plt.axvline(224, c='green')
plt.axvline(210, c='red')
```
